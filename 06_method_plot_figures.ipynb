```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.patheffects as pe

# Coordinates for the specified cities
cities = {
    "FAVR": [51.0447, -114.0719],
    "Brain IMPACT": [50.0447, -114.0719],
    "ADNI": [36.7783, -119.4179], # Davis, CA
    "AUCD": [35.7783, -119.4179], # Davis, CA
    "UMCC": [52.0907, 5.1214],
    "TRACE-VCI": [52.8676, 4.9041], # Amsterdam
    "ACE": [51.9225, 4.4792],
    "VASCAMY": [48.1351, 11.5820],
    "Harmonization": [1.3521, 103.8198],
    "PRODEM": [47.0707, 15.4395]
}

# Text offsets for labels
text_offsets = {
    "FAVR": (2, 2),
    "Brain IMPACT": (2, -2),
    "ADNI": (2, 2),
    "AUCD": (2, -2),
    "UMCC": (2, 3),
    "TRACE-VCI": (-24, 4),
    "ACE": (-10, -3),
    "VASCAMY": (1, 2),
    "Harmonization": (-35, 0),
    "PRODEM": (2, -3)
}

# Number of subjects per study
subjects = {
    "ADNI": 994,
    "TRACE-VCI": 821,
    "AUCD": 639,
    "PRODEM": 367,
    "UMCC": 227,
    "Harmonization": 207,
    "VASCAMY": 76,
    "Brain IMPACT": 53,
    "ACE": 52,
    "FAVR": 47
}

# Create a base map using Cartopy
fig, ax = plt.subplots(figsize=(12, 10), subplot_kw={'projection': ccrs.PlateCarree()})
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.stock_img()

pastel_cmap = plt.cm.get_cmap('tab10_r')

# Plot each city and adjust text position
for i, (city, coords) in enumerate(cities.items()):
    color = pastel_cmap(i / len(cities))
    ax.plot(coords[1], coords[0], marker='o', color=color, markersize=5, transform=ccrs.Geodetic())
    offset = text_offsets[city]
    ax.text(coords[1] + offset[0], coords[0] + offset[1], f"{city}", 
            fontsize=12, ha='left', va='center', color=color, 
            transform=ccrs.Geodetic(),
            path_effects=[pe.withStroke(linewidth=4, foreground="white")])

# Adjust the view to the relevant areas
ax.set_extent([-130, 120, -10, 60], crs=ccrs.PlateCarree())

# Show the plot
plt.savefig("../output/plots_pos/map_cartopy.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set the size of the heatmap
plt.figure(figsize=(5, 10))

# Generate random data for a 1000x500 heatmap
data = np.random.rand(100, 50)

# Create the heatmap using the magma colormap
ax = sns.heatmap(data, cmap='RdGy', cbar=False, annot=False)

# Remove the ticks
ax.set_xticks([])
ax.set_yticks([])

# Display the plot
plt.savefig("../output/plots_pos/heatmap_magma.png", dpi=300, bbox_inches='tight')
plt.show()
```
